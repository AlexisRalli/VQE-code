```python
def calc_exp_pauliword(count_dict):
    # takes correct part of bitstring when all lines measured

    n_zeros = 0
    n_ones = 0

    for bitstring in count_dict:
        
        parity_m_term = sum([int(bit) for bit in bitstring]) % 2

        if parity_m_term == 0:
            n_zeros += count_dict[bitstring]
        elif parity_m_term == 1:
            n_ones += count_dict[bitstring]
        else:
            raise ValueError('state {} not allowed'.format(measure_term))

    expectation_value = (n_zeros - n_ones) / (n_zeros + n_ones)

    return expectation_value
```

```python
def measure_state_vector_seq_rot(n_shots, Ansatz_state, n_qubits, S_index, anti_commuting_set, check_reduction=True):
    
    I_circuit = cirq.Circuit([cirq.I.on(cirq.LineQubit(i)) for i in range(n_qubits)])

    Reduction_circuit_circ, Ps, gamma_l = Build_reduction_circuit(anti_commuting_set, S_index,
                                                                  check_reduction=check_reduction)
    # change basis to measure
    circ_obj = Change_PauliWord_measurement_to_Z_basis(Ps)
    circuit = cirq.Circuit(cirq.decompose_once(
        (circ_obj(*cirq.LineQubit.range(circ_obj.num_qubits())))))
    
    full_circuit=cirq.Circuit(I_circuit, Reduction_circuit_circ, circuit)
    final_state = full_circuit.unitary().dot(Ansatz_state)
    
    qubit_list, _ = zip(*(list(*Ps.terms.keys())))
    
    meas_outputs = cirq.sample_state_vector(final_state, indices=qubit_list, repetitions=n_shots)
    
    meas_dict={}  
    for binary_out in meas_outputs:
        bin_arr_str = np.char.mod('%d', binary_out)
        bin_str = "".join(bin_arr_str)
        if bin_str in meas_dict.keys():
            meas_dict[bin_str]+=1
        else:
            meas_dict[bin_str]=1
    
    
    return meas_dict, gamma_l
```

```python
from scipy.sparse.linalg import expm
from openfermion import qubit_operator_sparse
from scipy.sparse import csr_matrix

def Build_R_seq_op_MAT(anti_commuting_set, S_index,n_qubits, check_reduction=False):
    """
    Function to build R_S (make up of all R_SK terms)

    Args:
        anti_commuting_set(list): list of anti commuting QubitOperators
        S_index(int): index for Ps in anti_commuting_set list
        check_reduction (optional, bool): use linear algebra to check that 𝑅s† 𝐻s 𝑅s == 𝑃s
    returns:
        full_RS_circuit(cirq.Circuit): Q_circuit for R_s operator
        Ps (QubitOperator): Pauli_S operator with cofactor of 1!
        gamma_l (float): normalization term

    """
    X_sk_theta_sk_list, full_normalised_set, Ps, gamma_l = Get_Xsk_op_list(anti_commuting_set, S_index)

    seq_rot_matrices = []
    circuit_list=[]
    for X_sk_Op, theta_sk in X_sk_theta_sk_list:
#         pauliword_X_sk = list(X_sk_Op.terms.keys())[0]
#         const_X_sk = list(X_sk_Op.terms.values())[0]
        
        from openfermion import qubit_operator_sparse
        X_sk_mat = qubit_operator_sparse(-1j*X_sk_Op, n_qubits=n_qubits)
        
        Rl_matrix = expm(X_sk_mat*theta_sk/2)
        seq_rot_matrices.append(Rl_matrix)
        
#         full_exp_circ_obj = full_exponentiated_PauliWord_circuit(QubitOperator(pauliword_X_sk, -1j),
#                                                                  theta_sk / 2 * const_X_sk)

#         circuit = cirq.Circuit(
#             cirq.decompose_once((full_exp_circ_obj(*cirq.LineQubit.range(full_exp_circ_obj.num_qubits())))))
        
#         circuit.append(cirq.Circuit([cirq.I.on(cirq.LineQubit(i)) for i in range (12)]))
        
#         print(np.allclose(Rl_matrix.todense(), circuit.unitary()))
        
#         circuit_list.append(circuit)

#     full_RS_circuit = cirq.Circuit(circuit_list)
    Rl_full = reduce(np.dot, seq_rot_matrices[::-1])
#     print(np.allclose(Rl_full.todense(), full_RS_circuit.unitary()))

    if check_reduction:

        H_S = QubitOperator()
        for QubitOp in full_normalised_set['PauliWords']:
            H_S += QubitOp
        from openfermion import qubit_operator_sparse
        H_S_matrix = qubit_operator_sparse(H_S)

        Ps_mat=qubit_operator_sparse(Ps, n_qubits=n_qubits)
        reduction_mat = Rl_full.dot(H_S_matrix.dot(Rl_full.conj().transpose()))

        if not (np.allclose(Ps_mat.todense(), reduction_mat.todense())):
            print('reduction circuit incorrect...   𝑅s 𝐻s 𝑅s† != 𝑃s')

    return Rl_full, Ps, gamma_l
```

```python
def measure_state_vector_LCU(n_shots, Ansatz_state, n_system_qubits, N_index, anti_commuting_set):
    
    I_circuit = cirq.Circuit([cirq.I.on(cirq.LineQubit(i)) for i in range(n_system_qubits)])

    R_uncorrected, Pn, gamma_l = Get_R_op_list(anti_commuting_set, N_index)
    R_corrected_Op_list, R_corr_list, ancilla_amplitudes, l1 = absorb_complex_phases(R_uncorrected)
    
    n_ancilla_qubits = int(np.ceil(np.log2(len(ancilla_amplitudes))))
    
    #  system kron ancilla!
    input_state = np.kron(Ansatz_state, reduce(np.kron, [np.array([[1],[0]]) for _ in range(n_ancilla_qubits)]))
    
    R_circ_obj = LCU_R_gate(n_ancilla_qubits, n_system_qubits, R_corrected_Op_list, R_corr_list, Pn)
    R_circ_circ = cirq.Circuit(
        cirq.decompose_once((R_circ_obj(*cirq.LineQubit.range(R_circ_obj.num_qubits())))))
    
    change_to_Z_basis_obj = Change_PauliWord_measurement_to_Z_basis(Pn)
    change_to_Z_basis_circ = cirq.Circuit(
        cirq.decompose_once((change_to_Z_basis_obj(*cirq.LineQubit.range(change_to_Z_basis_obj.num_qubits())))))

    
    ancilla_obj = prepare_arb_state(ancilla_amplitudes, n_system_qubits)
    ancilla_circ = ancilla_obj.Get_state_prep_Circuit()
    
    
    full_circuit=cirq.Circuit(I_circuit, ancilla_circ, R_circ_circ, change_to_Z_basis_circ, list(ancilla_circ.all_operations())[::-1])

    final_state = full_circuit.unitary().dot(input_state)
    
    qubit_list, _ = zip(*(list(*Pn.terms.keys())))
    ancilla_qubit_list = [i for i in range(n_system_qubits, n_ancilla_qubits+n_system_qubits)]
    
    # note need to include ancilla in indices!
    meas_outputs = cirq.sample_state_vector(final_state, indices=[*qubit_list, *ancilla_qubit_list], repetitions=n_shots)
    
    meas_dict={}  
    for binary_out in meas_outputs:
        bin_arr_str = np.char.mod('%d', binary_out)
        bin_str = "".join(bin_arr_str)
        if bin_str in meas_dict.keys():
            meas_dict[bin_str]+=1
        else:
            meas_dict[bin_str]=1
    
    
    return meas_dict, gamma_l
```
