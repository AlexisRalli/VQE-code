```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
## Adding nodes
G.add_node(1)
G.add_node(2)
print(G.nodes)

G.add_node(3)
print(G.nodes)
```

```python
# adding list of nodes
G.add_nodes_from([4,5,6])
print(G.nodes)
```

```python
# now add edges between these nodes!
G.add_edge(1,2)

G.add_edge(1,3)

G.add_edge(4,6)

G.add_edge(5,4)

G.add_edge(2,3)

G.add_edge(2,6)
```

```python
# To see these edges type:
print(G.edges())
```

```python
# let us visualize the graph!
import matplotlib.pyplot as plt

nx.draw(G)
plt.show()
```

```python
# but we don't know what node is what

# to display labels:

nx.draw(G, with_labels=1)
plt.show()
```

NetworkX has lots of built in functions

such as `complete_graph`

```python
# this function builds a complete graph on 10 vertices
Z = nx.complete_graph(10)
# aka all possible edges between 10 nodes!
# e.g. all possible friendships between 10 friends (= 45)
```

```python
# Checking

# should have 10 nodes
print(Z.nodes())
print('No. of nodes: ', len(Z.nodes()))

print('')
# should have 45 edges
print(Z.edges())
print('No. of edges: ', len(list(Z.edges())))
```

```python
# there is actually a shortcut for the above size...

# order gives number of nodes
print(Z.order())

# size gives number of edges in graph
print(Z.size())
```

```python
# visualising the graph!
nx.draw(Z, with_labels=1)
plt.show()

# can see it is a complete graph
# (every nodes connected to every other node)
```

# Graph colouring!

```python
GG = nx.cycle_graph(4)
x = nx.coloring.greedy_color(GG, strategy='largest_first')

## Returns
# A dictionary with keys representing nodes and values representing
# corresponding coloring.
print(x)
```

```python
pos=nx.circular_layout(GG)
nx.draw(G, pos, with_labels=1, node_color = ['r','b'])
```
